## <br><br><font color = red>1. Importe as devidas bibliotecas.</font>

```python
import pandas as pd
import numpy as np
from IPython.display import display
```

# <br><br><font color = blue>2. Abra o arquivo.</font>

```python
notas = pd.read_csv("notas2.csv", sep=";")
display(notas)

#pd.read_csv = Função para a abrir o arquivo.csv
#notas2.csv  = É o nome do arquivo que eu quero abrir
```

### <br><br><font color = green>3. Substitua valores `NaN`.</font>

```python
notas.fillna(0.0, inplace=True)
display(notas)
```

## <br><br><font color = orange>4. Substitua notas maiores do que 10 da coluna "Listas" por 10.</font>

```python
for i in list(notas.index.values): #Para "i" indo do primeiro ao último index da tabela notas
    if notas.at[i, "Listas"] > 10.0:
        notas.at[i, "Listas"] = 10.0
display(notas)
```

## <br><br><font color = plate>5. Preencha a coluna "Provas" corretamente com a conta <br><br>`(2*(nota da prova 1) + 3*(nota da prova 2))/5`.</font>

```python
notas["Provas"] = ((2 * notas["Prova 1"]) + (3 * notas["Prova 2"])) /5
display(notas)
```

### <br><br><font color = violet>6. Crie uma coluna nova de rótulo "Nota antes do exame" e a preencha com <br>a nota de cada aluno antes do exame. Essa nota $M$ é calculada da seguinte<br> forma, onde $P$ é a nota das provas e $L$ é a nota das listas:
$$\begin{array}{c}
M_1 = (7\times P + 3\times L)/10\\
M_2 = (3\times P\times L)/(P + 2\times L)\\
M = \max(M1, M2)\\
\end{array}$$</font>

```python
#Aqui eu estou inicializando a coluna "Nota antes do exame" 
notas["Nota antes do exame"] = 0.0
display(notas)
```

```python
#M1=(7×P+3×L)/10
#M2=(3×P×L)/(P+2×L)
#M=max(M1,M2)

#M1 = (7 × Provas + 3 × Listas) / 10
#M2 = (3 × Provas × Listas) / (Provas + 2 × Listas)
#M = max(M1,M2)

#"i" será a quantidade de vezes que o laço "FOR" se repetirá
for i in list(notas.index.values): #Para "i" indo do primeiro ao último index da tabela notas
    if notas.at[i, "Provas"] != 0 and notas.at[i, "Listas"] != 0: 
                
        m1 = (7 * notas.at[i, "Provas"] + 3 * notas.at[i, "Listas"]) / 10
        
        m2 = (3 * notas.at[i, "Provas"] * notas.at[i, "Listas"]) / (notas.at[i, "Provas"] + 2 * notas.at[i, "Listas"]) 
        
        m = max(m1, m2)
        notas.at[i, "Nota antes do exame"] = m

display(notas)
```

## <br><br><font color  = #009955>7. Crie uma coluna nova de rótulo "Nota final" e a preencha com a nota final de cada aluno após o exame. Essa nota $F$ é calculada da seguinte forma, onde $M$ é a nota antes do exame e $E$ é a nota do exame:
$$\begin{array}{c}
F = \min(5, (M + E)/2), \mbox{ se } $E > 0$ \\
F = M, \mbox{ caso contrário }\\
\end{array}$$</font>

```python
#Aqui eu estou inicializando a coluna "Nota final"
notas["Nota final"] = 0.0  
display(notas)
```

```python
#F = min(5,(M + E) / 2), se $E > 0$
#F = M, caso contrário

#se $Exame > 0$ 
#    F = min(5,(Nota antes do exame + Exame) / 2)
#else
#    F = Nota antes do exame

#"i" será a quantidade de vezes que o laço "FOR" se repetirá
for i in list(notas.index.values): #Para "i" indo do primeiro ao último index da tabela notas
    if notas.at[i, "Exame"] > 0:
        f = min(5,(notas.at[i, "Nota antes do exame"] + notas.at[i, "Exame"]) / 2)
        notas.at[i, "Nota final"] = f
    else:
        f = notas.at[i, "Nota antes do exame"]
        notas.at[i, "Nota final"] = f
display(notas)
```

## <br><br><font color = #115522>8. Remova da tabela os alunos que ficaram com nota 0 em todas as atividades.</font>

```python
condicao = (notas["Prova 1"] <= 0.0) & (notas["Prova 2"] <= 0.0) & (notas["Provas"] <= 0.0) & (notas["Listas"] <= 0.0)
linhas = list(notas[condicao].index.values)
notas.drop(linhas, axis=0, inplace=True)

display(notas)
```

### <br><br><font color = #991132>9. Crie uma coluna nova de rótulo "Conceito final" e a preencha com o conceito final de cada aluno. Dada a nota final $F$ após o exame, o conceito do aluno é:
$$\begin{array}{cr}
A, & \mbox{ se } F \geq 8.5 \\
B, & \mbox{ se } 7.0 \geq F < 8.5 \\
C, & \mbox{ se } 6.0 \geq F < 7.0 \\
D, & \mbox{ se } 5.0 \geq F < 6.0 \\
F, & \mbox{ se } F < 5.0 \\
\end{array}$$</font>

```python
##Aqui eu estou inicializando a coluna "Conceito Final"
notas["Conceito Final"] = ""
display(notas)
```

```python
#"i" será a quantidade de vezes que o laço "FOR" se repetirá
for i in list(notas.index.values):#Para "i" indo do primeiro ao último index da tabela notas
    if notas.at[i, "Nota final"] < 5.0:
        notas.at[i, "Conceito Final"] = "F"
    elif notas.at[i, "Nota final"] < 6.0:
        notas.at[i, "Conceito Final"] = "D"
    elif notas.at[i, "Nota final"] < 7.0:
        notas.at[i, "Conceito Final"] = "C"
    elif notas.at[i, "Nota final"] < 8.5:
        notas.at[i, "Conceito Final"] = "B"
    else:
        notas.at[i, "Conceito Final"] = "A"
display(notas)
```

## <br><font color = magenta>10. Mostre os alunos que ficaram com conceito A.</font>

```python
condicao = notas["Conceito Final"] == "A"
display(notas[condicao])
```

## <br><br>11. Ordene a tabela por conceito final.

```python
notas.sort_values(by=["Conceito Final"], ascending = True, inplace = True)
display(notas)
```

### <br><font color = bold>12. Salve o conteúdo da tabela em um arquivo chamado "notas2_final.csv"</font>

```python
notas.to_csv("notas2_final.csv", sep=";")
```
