Vamos utilizar cinco arquivos nesse trabalho.

O arquivo pop_residente.csv descreve a nacionalidade dos indivíduos entrevistados.

O arquivo nivel_instrucao.csv descreve o nível de instrução dos indivíduos entrevistados.

O arquivo ocupacao.csv descreve a condição de atividade e ocupação dos indivíduos entrevistados.

O arquivo renda.csv descreve o rendimento mensal, em salários mánimos (SM), dos indivíduos
entrevistados com mais de 10 anos.

O arquivo agua.csv descreve a existência de água canalizada e a forma de abastecimento de água
nos domicílios dos indivíduos entrevistados.

```python
pop_residente = pd.read_csv("pop_residente.csv", sep=";")
nivel_instrucao = pd.read_csv("nivel_instrucao.csv", sep=";")
ocupacao = pd.read_csv("ocupacao.csv", sep=";")
renda = pd.read_csv("renda.csv", sep=";")
agua = pd.read_csv("agua.csv", sep=";")
```

```python
sul = ["Paraná", "Santa Catarina", "Rio Grande do Sul"]
norte = ["Acre", "Amapá", "Amazonas", "Pará", "Rondônia", "Roraima", "Tocantins"]
nordeste = ["Alagoas", "Bahia", "Ceará", "Maranhão", "Paraíba", "Pernambuco", "Piauí", "Rio Grande do Norte", "Sergipe"]
centrooeste = ["Goiás", "Mato Grosso", "Mato Grosso do Sul", "Distrito Federal"]
sudeste = ["Espírito Santo", "Minas Gerais", "São Paulo", "Rio de Janeiro"]
```

```python
agua["Região"] = ""
for i in list(agua.index.values):
    if agua.at[i, "Estado"] in sul:
        agua.at[i, "Região"] = "Sul"
    elif agua.at[i, "Estado"] in norte:
        agua.at[i, "Região"] = "Norte"
    elif agua.at[i, "Estado"] in nordeste:
        agua.at[i, "Região"] = "Nordeste"
    elif agua.at[i, "Estado"] in centrooeste:
        agua.at[i, "Região"] = "Centro-Oeste"
    else: #if agua.at[i, "Estado"] in sudeste:
        agua.at[i, "Região"] = "Sudeste"
        
ocupacao["Região"] = ""
for i in list(ocupacao.index.values):
    if ocupacao.at[i, "Estado"] in sul:
        ocupacao.at[i, "Região"] = "Sul"
    elif ocupacao.at[i, "Estado"] in norte:
        ocupacao.at[i, "Região"] = "Norte"
    elif ocupacao.at[i, "Estado"] in nordeste:
        ocupacao.at[i, "Região"] = "Nordeste"
    elif ocupacao.at[i, "Estado"] in centrooeste:
        ocupacao.at[i, "Região"] = "Centro-Oeste"
    else: #if ocupacao.at[i, "Estado"] in sudeste:
        ocupacao.at[i, "Região"] = "Sudeste"

pop_residente["Região"] = ""
for i in list(pop_residente.index.values):
    if pop_residente.at[i, "Estado"] in sul:
        pop_residente.at[i, "Região"] = "Sul"
    elif pop_residente.at[i, "Estado"] in norte:
        pop_residente.at[i, "Região"] = "Norte"
    elif pop_residente.at[i, "Estado"] in nordeste:
        pop_residente.at[i, "Região"] = "Nordeste"
    elif pop_residente.at[i, "Estado"] in centrooeste:
        pop_residente.at[i, "Região"] = "Centro-Oeste"
    else: #if pop_residente.at[i, "Estado"] in sudeste:
        pop_residente.at[i, "Região"] = "Sudeste"
        
nivel_instrucao["Região"] = ""
for i in list(nivel_instrucao.index.values):
    if nivel_instrucao.at[i, "Estado"] in sul:
        nivel_instrucao.at[i, "Região"] = "Sul"
    elif nivel_instrucao.at[i, "Estado"] in norte:
        nivel_instrucao.at[i, "Região"] = "Norte"
    elif nivel_instrucao.at[i, "Estado"] in nordeste:
        nivel_instrucao.at[i, "Região"] = "Nordeste"
    elif nivel_instrucao.at[i, "Estado"] in centrooeste:
        nivel_instrucao.at[i, "Região"] = "Centro-Oeste"
    else: #if nivel_instrucao.at[i, "Estado"] in sudeste:
        nivel_instrucao.at[i, "Região"] = "Sudeste"

renda["Região"] = ""
for i in list(renda.index.values):
    if renda.at[i, "Estado"] in sul:
        renda.at[i, "Região"] = "Sul"
    elif renda.at[i, "Estado"] in norte:
        renda.at[i, "Região"] = "Norte"
    elif renda.at[i, "Estado"] in nordeste:
        renda.at[i, "Região"] = "Nordeste"
    elif renda.at[i, "Estado"] in centrooeste:
        renda.at[i, "Região"] = "Centro-Oeste"
    else: #if renda.at[i, "Estado"] in sudeste:
        renda.at[i, "Região"] = "Sudeste"
```

Através do gráfico a seguir é possível visualizar a distribuição da população 
estrangeira ,por região, no território brasileiro, de acordo com os dados do Censo Demográfico 2010.

```python
pop_residente.groupby("Região").agg("sum").plot(y="Residente Estrangeira", kind="pie", figsize=(16,10), 
                                                autopct='%1.1f%%').legend(bbox_to_anchor=(1.0, 1))
```

Por meio do gráfico acima é possível perceber que a maior parcela da população estrangeira se encontra na região sudeste.

Já que foi constatado que a região sudeste é a que concentra o maior número de estrangeiros, agora veremos 
quais são os estados da região sudeste que concentram o maior e o menor número de estrangeiros, respectivamente.

```python
condicao = ((pop_residente["Estado"] == "São Paulo") | (pop_residente["Estado"] == "Rio de Janeiro") | 
            (pop_residente["Estado"] == "Minas Gerais") | (pop_residente["Estado"] == "Espírito Santo"))
populacao_agrupada = pop_residente[condicao].groupby("Estado").agg("sum")
populacao_agrupada.plot(y=["Residente Estrangeira"], title = "População Estrangeira na região Sudeste", kind="pie", figsize=(16,10), autopct='%1.1f%%').legend(bbox_to_anchor=(1.0, 1))
```

A seguir faremos uma comparação entre as regiões do Brasil, a respeito do nível de instrução.

```python
nivel_instrucao.groupby("Região").agg("sum").plot(y=["Sem instrução e fundamental incompleto", 
                                                 "Fundamental completo e médio incompleto",
                                                 "Médio completo e superior incompleto", "Superior completo"], 
                                                  kind="barh", figsize=(16,10) ,logx = True).legend(bbox_to_anchor=(1.0, 1))
```

Aqui está sendo feita uma comparação entre o nível de instrução dos estados da região norte com os da região sudeste do Brasil (que é considearada, de acordo com dados do IBGE a mais rica do país).

```python
condicao = ((nivel_instrucao["Estado"] == "São Paulo") | (nivel_instrucao["Estado"] == "Rio de Janeiro") | 
           (nivel_instrucao["Estado"] == "Minas Gerais") | (nivel_instrucao["Estado"] == "Espírito Santo") |
           (nivel_instrucao["Estado"] == "Rondônia") | (nivel_instrucao["Estado"] == "Acre") | 
           (nivel_instrucao["Estado"] == "Amazonas") | (nivel_instrucao["Estado"] == "Para") |
           (nivel_instrucao["Estado"] == "Roraima") | (nivel_instrucao["Estado"] == "Amapá") |
           (nivel_instrucao["Estado"] == "Tocantins"))

nivel_instrucao_agrupada = nivel_instrucao[condicao].groupby("Estado").agg("sum")
nivel_instrucao_agrupada.plot(y=["Sem instrução e fundamental incompleto", "Fundamental completo e médio incompleto", 
                                  "Médio completo e superior incompleto", "Superior completo"],
                   kind="barh", logx=True, figsize=(16,11)).legend(bbox_to_anchor=(1.0, 0.12))
```
